```python
# RAW h5 file viewer:

pgroup = 22250
runnumber = 192
f = f'/sf/alvra/data/p{pgroup}/raw/run{runnumber:04d}*/data/*'
```

```python
import h5py
import ipywidgets as widgets
#Open the H5 file in read mode
FILELIST = widgets.Select(
    options=glob.glob(f),
    rows=60,
    disabled=False,
    layout=widgets.Layout(width='30%')
)

KEYLIST = widgets.Select(
    options=[],
    rows=60,
    disabled=False,
    layout=widgets.Layout(width='20%')
)

def _get_key_list(change):
    f = FILELIST.value
    with h5py.File(f, 'r') as file:
        fl=file.keys()
        KEYLIST.options=fl
        KEYLIST.value=fl[0]

output = widgets.Output()

LIST.options = list(file.keys())

def on_select_change(change):
    f = FILELIST.value
    _item = KEYLIST.value
    output.clear_output()
    print(f)
    
    with output:
        with h5py.File(f, 'r') as file:
            for k in file[_item].keys():
                print(k)
                print(file[_item][k][:])
FILELIST.observe(_get_key_list, 'value')
KEYLIST.observe(on_select_change, 'value')
widgets.HBox([FILELIST, KEYLIST, output])
```
