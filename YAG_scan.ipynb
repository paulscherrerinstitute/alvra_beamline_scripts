```python
plt.figure()
plt.title('{} -- Runs 131 to 134'.format(pgroup))
plt.scatter(energy, delay)
plt.plot(energy, m*energy+b, linestyle ='--', label = '{:.5f} fs/eV'.format(m))
plt.grid()
plt.xlabel('DCM energy (eV)')
plt.ylabel('Edge position (fs)')
plt.legend(loc='best')
plt.show()
```
